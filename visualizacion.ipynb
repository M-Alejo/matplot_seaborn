### Pyplot básico


````python
# crear variables x e y
x = np.linspace(0,5,11)
y = x**2
````

````python
# crear grafica
plt.plot(x,y)
plt.show()
````

````python
# crear modificar color de graficas y marcadores
plt.plot(x, y, 'rx')
plt.show()
````

````python
# crear modificar color de graficas y marcadores y agregar linea continua
plt.plot(x, y, 'rs-') # r: color rojo; s:forma cuadrados; -:linea continua
plt.show()
````

````python
# crear modificar color de graficas y marcadores y agregar linea punteada
plt.plot(x, y, 'yD:') # y: color amarillo; D:forma diamante; ::linea punteada
plt.show()
````

````python
# crear histograma
plt.hist(x)
plt.show()
````

````python
# crear gráfica de pie  torta
plt.pie(x)
plt.show()
````

````python
# crear gráfica de scatter, para relacionar dos variables
plt.scatter(x, y)
plt.show()
````

````python
# crear gráfica boxplot de una variable
plt.boxplot(x)
plt.show()
````

### Método orientado a objeto en Matplotlib
Aunque tiene un poco más de código se puede personalizar más.
Pasos:
* Definir lienzo con: nombre_lienzo =  plt.figure()
* Agregar axes o cuadro donde va la primera gráfica con: nombre_axes = fig.add_axes([posx, posy, sizex, sizey])
* Agregar gráfica en axes o cuadro creado con nombre_axes.plot()
* Agregar a cada axes la configuración deseada
* mostrar el lienzo con nombre_lienzo.show()

![Método orientado a objetos](./metodo_oop.png) ![Ejemplo](ejemplo_oop.png) ![Estructura general de una figura](./estructura_oop.png) ![Personalizar](./personalizable_oop2.png)

Figuras tomadas de [Matplotlib](https://matplotlib.org/stable/gallery/showcase/anatomy.html)

### Subplots con método orientado a objetos

````python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas, si se pasa sin argumentos crea un solo axes
fig, axes = plt.subplots()
axes.plot(x,y, 'b')

#se muestra el lienzo completo
fig.show()
````

````python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array
# se crea el lienzo fig, con los axes como array
fig, axes = plt.subplots(nrows=1, ncols=2)

#grafico en el primer axes, el primer axes es 0
axes[0].plot(x,y, 'b')

#grafico en el segundo axes
axes[1].plot(y,x, 'r:')

#se muestra el lienzo completo
fig.show()
````

````python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array

# se crea el lienzo fig, los axes se asignan con una tupla
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

#grafico en ax1
ax1.plot(x,y, 'b')

#grafico en ax2
ax2.plot(y,x, 'r:')

#se muestra el lienzo completo
fig.show()
````

````python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array

# se crea el lienzo fig, con ocho axes organizados en dos filas y cuatro columnas
fig, axes = plt.subplots(nrows=2, ncols=4)

#grafico en axes[0,0]: [nfila, ncolummna], el axes de la primera fila y la primera columna
axes[0,0].plot(x,y, 'b')

#grafico en axes[0,1] : [nfila, ncolumna], el axes de la primera fila y la segunda columna
axes[0,1].plot(y,x, 'r:')

#se muestra el lienzo completo
fig.show()
````

````python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array

# se crea el lienzo fig, con ocho axes organizados en dos filas y cuatro columnas, con nombre propios dados por tupla
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(nrows=2, ncols=4) # despues de fig van los nombres de los axes ((axes de primera fila),(axes de segundafila))

# grafico en ax1, el axes de la primera fila y la primera columna
ax1.plot(x,y, 'b')

# grafico en ax2, el axes de la primera fila y la segunda columna
ax2.plot(x**3,y, 'b--')

# grafico en ax3, el axes de la primera fila y la tercera columna
ax3.plot(x,y**3, 'y:')

# grafico en ax4, el axes de la primera fila y la cuarta columna
ax4.plot(x,y**2, 'bs-')

# grafico en ax8, el axes de la segunda fila y ultima columna
ax8.plot(y,x, 'r:')

# se ajusta la distancia entre figuras
fig.tight_layout()

#se muestra el lienzo completo
fig.show()
````

### Leyendas, etiquetas, títulos y tamaño

````python
# se crea lienzo y axes, en la creacion de la figura se define el tamano del lienzo con figsize=(xdim, ydim)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 5))

# se grafica en el axes[0]
axes[0].plot(x, y, label='línea x') # agrega grafica con nombre de linea en label='nombre de linea'
axes[0].set_title('Relación X vs Y') # se agrega titulo a grafica
axes[0].set_xlabel('x') # se agrega titulo de eje x
axes[0].set_ylabel('y') # se agrega titulo de eje y
axes[0].legend() # para mostrar nombre de lineas o leyenda de la grafica

# grafica en el axes[1]
axes[1].plot(y, x, label='línea y')
axes[1].set_title('Relación Y vs X') # se agrega titulo a grafica
axes[1].set_xlabel('y') # se agrega titulo de eje x
axes[1].set_ylabel('x') # se agrega titulo de eje y
axes[1].legend() # para mostrar nombre de lineas o leyenda de la grafica

# se muestra la figura
fig.show()
````

#### Manejar estilos de línea con la notación MATLAB
La notación MATLAB es la que va entre comillas sencillas por ejemplo 'yo-' que significa color amarillo, marcador de punto y linea que une marcadores

````python
# crear figura
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x, x+1, 'yo-') # grafica color amarilla con marcador de punto y linea
ax.plot(x, x+2, 'bD-') # grafica color azul con marcador de diamante y linea
ax.plot(x, x+3, 'r:') # grafica colo rojo linea punteada sin marcador
ax.plot(x, x+4, 'o') # grafica con marcador de punto sin linea

# se muestra la figura
fig.show()
````

#### Manejar estilos con propiedades de MatplotLib
Con Matplotlib se accede a la propiedad por el nombre


````python
# crear figura
fig, ax = plt.subplots(figsize=(8,8))

# grafica 1
ax.plot(x, x+1, # datos que grafica
        color='green', # color de linea
        alpha=0.5, # transparencia de linea
        linewidth=1, # grosor de linea
        linestyle='dashed', # estio de linea
        marker='o', # marcador redondo
        markersize=10, # tamano de marcador
        markerfacecolor='blue' # color de marcador
        ) 

# grafica 2
ax.plot(x, x+2, # datos que grafica
        color='blue', # color de linea
        alpha=0.7, # transparencia de linea
        linewidth=2, # grosor de linea
        linestyle='-', # estio de linea continua
        marker='X' # marcador de x
        )

# grafica 3
ax.plot(x, x+3, # datos que grafica
        color='black', # color de linea
        alpha=1, # transparencia de linea
        linewidth=3, # grosor de linea
        linestyle=':', # estio de linea punteada
        marker='s' # marcador de cuadrado
        ) 

# grafica 4
ax.plot(x, x+4, # datos que grafica
        color='green', # color de linea
        alpha=0.5, # transparencia de linea
        linewidth=4, # grosor de linea
        linestyle='--', # estio de linea
        marker='D' # marcador de diamante
        ) 

# se muestra la figura
fig.show()
````

### Bar plot con pyplot, para variables categoricas

````python
# definir variables categoricas
paises = ['INDIA', 'JAPON', 'MEXICO', 'COLOMBIA', 'ALEMANIA']
poblacion = [1000, 800, 900, 1200, 300]
````

````python
# grafica de barras con pyplot
plt.bar(paises, poblacion)
plt.show()
````

````python
# grafica de barras modificada con pyplot

plt.bar(paises, # categorias
        poblacion, # valores
        width=0.5, #ancho de barras
        color=['blue', 'green'] # poniendo dos colores python los alterna
        )

# cambiar nombres de eje x
plt.xticks(
    np.arange(5), # se pone la cantidad de categorias que tenga, en este caso son cinco paises
    ('India', 'Japón', 'México', 'Colombia', 'Alemania'), #nombres que quiero poner, deben estar en el mismo orden de las categorias iniciales
    rotation=45 # rotar los nombres de categorias en grados
)

plt.show()
````

````python
# grafica de barras horizontal
plt.barh(paises, poblacion)
plt.show()
````

````python
# tambien se pueden cambiar colores igual que con las barras verticales

plt.barh(paises, # categorias
        poblacion, # valores
        color=['blue', 'green'] # poniendo dos colores python los alterna
        )

# cambiar nombres de eje y
plt.yticks(
    np.arange(5), # se pone la cantidad de categorias que tenga, en este caso son cinco paises
    ('India', 'Japón', 'México', 'Colombia', 'Alemania') #nombres que quiero poner, deben estar en el mismo orden de las categorias iniciales
)

plt.show()
````

### Otros gráficos

````python
# se crea un conjunto de 100 datos aleatorio con valores enteros entre 1 y 50
datos = np.random.randint(1, 50, 100)
print(datos)
````

#### Histograma con pyplot

````python
# grafica de histograma, util para ver distribucion de frecuencias
plt.hist(datos, #datos usados para graficar
         bins='sturges', # numero de barras, por defecto viene en 10, se puede definir con un metodo: 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', or 'sqrt'.
         width=3 #ancho de las barras
         )

plt.show()
````

#### Boxplot con pyplot

````python
# hacer grafica de boxplot con pyplot

plt.boxplot(datos, # datos con los que se hace el boxplot
            vert=False, # para ponerlo horizontal, s esta en True el boxplot sale vertical
            patch_artist=True, # para rellenar la caja, si esta en False no viene rellena la caja
            notch=True, # para angostar en el percentil 50
            showfliers=True # para ocultar mostrar valores atipicos, si esta en false los oculta
            )
# mostrar la figura
plt.show()
````

#### Scatter plotcon pyplot

````python
# crear datos aleatorios
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30*  np.random.rand(n)) ** 2
colors = np.random.rand(n)
````

````python
# grafica de scatter con pyplot
plt.scatter(x, y)
plt.show()
````

````python
# grafica de scatter con pyplot, agregando otras caracteristicas
plt.scatter(x, # datos eje x
            y, # datos eje y
            s=area, # tamano de puntos, se peude asociar a otra variable
            c = colors, # colores de puntos
            marker='X', # tipo de marcador
            alpha=0.5 # transparencia, se mostraria mas oscuro en donde haya mas sobreposicion de datos
            )

plt.show()
````

````python
# importar libreria de seaborn y matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
````

### Set o configuración básica de las gráficas que se van a generar
* Seaborn tiene cinco temas: darkgrid, whitegrid, dark, white, ticks. Más información [Aquí](https://seaborn.pydata.org/tutorial/aesthetics.html?highlight=sns+set)
* Seaborn tiene seis paletas de colores: pastel, muted, bright, deep, colorblind, dark. Más información [Aquí](https://seaborn.pydata.org/tutorial/color_palettes.html?highlight=palette)

````python
#tema darkgrid
sns.set(style='darkgrid')

#grafica
sns.barplot(x=['A', 'B', 'C'],
            y=[1, 3, 2])
plt.show()
````

````python
#tema whitegrid
sns.set(style='whitegrid')

#grafica
sns.barplot(x=['A', 'B', 'C'],
            y=[1, 3, 2])
plt.show()
````

````python
#tema white
sns.set(style='white')

#grafica
sns.barplot(x=['A', 'B', 'C'],
            y=[1, 3, 2])
plt.show()
````

````python
#tema dark
sns.set(style='dark')

#grafica
sns.barplot(x=['A', 'B', 'C'],
            y=[1, 3, 2])
plt.show()
````

````python
#tema ticks
sns.set(style='ticks')

#grafica
sns.barplot(x=['A', 'B', 'C'],
            y=[1, 3, 2])
plt.show()
````

````python
# con set se puden cambiar tambien la paleta de colores, tamanp de letra
sns.set(style='darkgrid', # color de fondo
        palette='dark', # colores de barras
        font_scale=2 # tamano de fuente
        )

# grafica
sns.barplot(x=['A', 'B', 'C'],
            y=[1, 3, 2])
plt.show()
````
