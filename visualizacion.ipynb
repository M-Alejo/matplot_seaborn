```python
# importar librerias
import matplotlib.pyplot as plt
import numpy as np
```

### Pyplot básico


```python
# crear variables x e y
x = np.linspace(0,5,11)
y = x**2
```

```python
# crear grafica
plt.plot(x,y)
plt.show()
```

```python
# crear modificar color de graficas y marcadores
plt.plot(x, y, 'rx')
plt.show()
```

```python
# crear modificar color de graficas y marcadores y agregar linea continua
plt.plot(x, y, 'rs-') # r: color rojo; s:forma cuadrados; -:linea continua
plt.show()
```

```python
# crear modificar color de graficas y marcadores y agregar linea punteada
plt.plot(x, y, 'yD:') # y: color amarillo; D:forma diamante; ::linea punteada
plt.show()
```

```python
# crear histograma
plt.hist(x)
plt.show()
```

```python
# crear gráfica de pie  torta
plt.pie(x)
plt.show()
```

```python
# crear gráfica de scatter, para relacionar dos variables
plt.scatter(x, y)
plt.show()
```

```python
# crear gráfica boxplot de una variable
plt.boxplot(x)
plt.show()
```

### Manejo de Subplot

```python
# crear subplot parametros (nfilas, ncolumnas, indice actual)
plt.subplot(1, 2, 1) # una fila dos columnas cajita uno
plt.plot(x,y)

plt.subplot(1, 2, 2) # una fila dos columnas cajita dos
plt.hist(y)

plt.show()
```

```python
# crear subplot parametros (nfilas, ncolumnas, indice actual)
plt.subplot(1, 2, 1) # una fila dos columnas cajita uno
plt.plot(x,y, 'r--') # primera grafica en la caja uno
plt.plot(y,x, 'b:') # segunda grafica en la caja uno

plt.subplot(1, 2, 2) # una fila dos columnas cajita dos
plt.pie(y)

plt.show()
```

```python
# crear subplot parametros (nfilas, ncolumnas, indice actual)
plt.subplot(2, 1, 1) # una fila dos columnas cajita uno
plt.plot(x,y, 'r--') # primera grafica en la caja uno
plt.plot(y,x, 'b:') # segunda grafica en la caja uno

plt.subplot(2, 1, 2) # una fila dos columnas cajita dos
plt.pie(y)

plt.show()
```

### Método orientado a objeto en Matplotlib
Aunque tiene un poco más de código se puede personalizar más.
Pasos:
* Definir lienzo con: nombre_lienzo =  plt.figure()
* Agregar axes o cuadro donde va la primera gráfica con: nombre_axes = fig.add_axes([posx, posy, sizex, sizey])
* Agregar gráfica en axes o cuadro creado con nombre_axes.plot()
* Agregar a cada axes la configuración deseada
* mostrar el lienzo con nombre_lienzo.show()

![Método orientado a objetos](./metodo_oop.png) ![Ejemplo](ejemplo_oop.png) ![Estructura general de una figura](./estructura_oop.png) ![Personalizar](./personalizable_oop.png)

```python
# usando metodo orientago a objetos
fig = plt.figure() # lienzo sobre el que se hacen las figuras
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9]) # [posx, posy, xsize, ysize] los dos primeros indican la posicion en el lienzo, los otros dos el tamano
axes.plot(x, y, 'b')
fig.show()
```

```python
# usando metodo orientago a objetos

# definir lienzo
fig = plt.figure() # lienzo sobre el que se hacen las figuras

# definir axes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) # [posx, posy, xsize, ysize] los dos primeros indican la posicion en el lienzo, los otros dos el tamano
axes2 = fig.add_axes([0.1, 0.1, 0.4, 0.3]) # [posx, posy, xsize, ysize] los dos primeros indican la posicion en el lienzo, los otros dos el tamano

# agregar graficas en cada axes
axes.plot(x, y, 'b')
axes2.plot(y, x, 'r--')

#mostrar el lienzo
fig.show()
```

```python
# usando metodo orientago a objetos

# definir lienzo
fig = plt.figure() # lienzo sobre el que se hacen las figuras

# definir axes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9]) # [posx, posy, xsize, ysize] los dos primeros indican la posicion en el lienzo, los otros dos el tamano
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3]) # [posx, posy, xsize, ysize] los dos primeros indican la posicion en el lienzo, los otros dos el tamano

# agregar graficas en cada axes
axes.plot(x, y, 'b')
axes2.plot(y, x, 'r--')

# personalizar axes
axes2.set_facecolor('gray')

#mostrar el lienzo
fig.show()
```

### Subplots con método orientado a objetos

```python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas, si se pasa sin argumentos crea un solo axes
fig, axes = plt.subplots()
axes.plot(x,y, 'b')

#se muestra el lienzo completo
fig.show()
```

```python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array
# se crea el lienzo fig, con los axes como array
fig, axes = plt.subplots(nrows=1, ncols=2)

#grafico en el primer axes, el primer axes es 0
axes[0].plot(x,y, 'b')

#grafico en el segundo axes
axes[1].plot(y,x, 'r:')

#se muestra el lienzo completo
fig.show()
```

```python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array

# se crea el lienzo fig, los axes se asignan con una tupla
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

#grafico en ax1
ax1.plot(x,y, 'b')

#grafico en ax2
ax2.plot(y,x, 'r:')

#se muestra el lienzo completo
fig.show()
```

```python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array

# se crea el lienzo fig, con ocho axes organizados en dos filas y cuatro columnas
fig, axes = plt.subplots(nrows=2, ncols=4)

#grafico en axes[0,0]: [nfila, ncolummna], el axes de la primera fila y la primera columna
axes[0,0].plot(x,y, 'b')

#grafico en axes[0,1] : [nfila, ncolumna], el axes de la primera fila y la segunda columna
axes[0,1].plot(y,x, 'r:')

#se muestra el lienzo completo
fig.show()
```

```python
# subplots sirve para agregar varios axes en una figura nrows, ncols define cuantas axes es un array

# se crea el lienzo fig, con ocho axes organizados en dos filas y cuatro columnas, con nombre propios dados por tupla
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(nrows=2, ncols=4) # despues de fig van los nombres de los axes ((axes de primera fila),(axes de segundafila))

# grafico en ax1, el axes de la primera fila y la primera columna
ax1.plot(x,y, 'b')

# grafico en ax2, el axes de la primera fila y la segunda columna
ax2.plot(x**3,y, 'b--')

# grafico en ax3, el axes de la primera fila y la tercera columna
ax3.plot(x,y**3, 'y:')

# grafico en ax4, el axes de la primera fila y la cuarta columna
ax4.plot(x,y**2, 'bs-')

# grafico en ax8, el axes de la segunda fila y ultima columna
ax8.plot(y,x, 'r:')

# se ajusta la distancia entre figuras
fig.tight_layout()

#se muestra el lienzo completo
fig.show()
```

### Leyendas, etiquetas, títulos y tamaño

```python
# se crea lienzo y axes, en la creacion de la figura se define el tamano del lienzo con figsize=(xdim, ydim)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 5))

# se grafica en el axes[0]
axes[0].plot(x, y, label='línea x') # agrega grafica con nombre de linea en label='nombre de linea'
axes[0].set_title('Relación X vs Y') # se agrega titulo a grafica
axes[0].set_xlabel('x') # se agrega titulo de eje x
axes[0].set_ylabel('y') # se agrega titulo de eje y
axes[0].legend() # para mostrar nombre de lineas o leyenda de la grafica

# grafica en el axes[1]
axes[1].plot(y, x, label='línea y')
axes[1].set_title('Relación Y vs X') # se agrega titulo a grafica
axes[1].set_xlabel('y') # se agrega titulo de eje x
axes[1].set_ylabel('x') # se agrega titulo de eje y
axes[1].legend() # para mostrar nombre de lineas o leyenda de la grafica

# se muestra la figura
fig.show()
```
